Завдання 1: На вхід функція приймає рядок s. Функція повинна повернути True якщо рядок є паліндромом та False – в протилежному випадку.

Наприклад:

*   `is_palindrome(s='anade') -> False`
*   `is_palindrome(s='ana') -> True`




````python
def is_palindrome(s):
    # Усуваємо всі нелітерні символи з рядка
    s = ''.join(c for c in s if c.isalpha())
    
    # Перевіряємо, чи є рядки однаковими, коли ми їх обернемо
    return s == s[::-1]

# Приклади використання
print(is_palindrome('anade'))  # False
print(is_palindrome('ana'))  # True
````

Завдання 2: Розв’язати квадратне рівняння $ax^2+bx + c = 0$.  Напишіть функцію,яка приймає параметри квадратного рівняння, а повертає одне рішення або декілька.

Наприклад:  


```
solve_quad(a=1, b=-2, c=1) -> 1  
solve_quad(a=1, b=0, c=-1) -> (-1, 1)  
solve_quad(a=0, b=2, c=-1) -> 0.5
```







````python
import cmath

def solve_quad(a, b, c):
    # Вираховуємо дискримінант
    discriminant = b**2 - 4*a*c
    
    # Якщо дискримінант менше 0, то рівняння не має дійсних рішень
    if discriminant < 0:
        return None
    
    # Якщо дискримінант дорівнює 0, то рівняння має одне дійсне рішення
    elif discriminant == 0:
        return -b / (2*a)
    
    # Якщо дискримінант більше 0, то рівняння має дві дійсні рішення
    else:
        # Вираховуємо дві рішення
        root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        root2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        
        # Повертаємо рішення
        return (root1, root2)

# Використання функції
print(solve_quad(a=1, b=-2, c=1)) 
print(solve_quad(a=1, b=0, c=-1))  
print(solve_quad(a=10, b=2, c=-1))  
````

Завдання 3: Реалізація функції merge. На вхід функція приймає 2 відсортованих масиви - `a` і `b`. Функція повинна повернути один відсортований масив, який містить всі елементи масивів `a` і `b`.

Наприклад: `merge([1, 3, 3, 5], [3, 5, 5, 7]) -> [1, 3, 3, 3, 5, 5, 5, 7]`

````python
def merge(a, b):
    # Базовий випадок: коли один з масивів пустий
    if not a:
        return b
    if not b:
        return a
    
    # Рекурсивний випадок: коли обидва масиви не пусті
    # Вибираємо найменший елемент з двох масивів
    if a[0] < b[0]:
        merged = [a[0]]
        a = a[1:]
    else:
        merged = [b[0]]
        b = b[1:]
    
    # Об'єднуємо залишки двох масивів
    return merged + merge(a, b)

# Використання функції
print(merge([1, 3, 3, 5], [3, 5, 5, 7]))  # [1, 3, 3, 3, 5, 5, 5, 7]
````

Завдання 4: Реалізація алгоритму [MergeSort](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC)..

Наприклад: `MergeSort([3, 1, 3, 5]) -> [1, 3, 3, 5]`

````python
def merge_sort(arr):
    # Базовий випадок: коли масив складається з одного або менше елементів
    if len(arr) <= 1:
        return arr
    
    # Розділяємо масив на дві частини
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    # Рекурсивно сортуємо обидві частини
    left = merge_sort(left)
    right = merge_sort(right)
    
    # Об'єднуємо сортовані частини
    return merge(left, right)

# Використання функції
print(merge_sort([3, 1, 3, 5]))  # [1, 3, 3, 5]
````

Завдання 6: Написати функцію яка на вхід приймає 3 масиви – `X, Y, Z`. функція повертатиме `True` у випадку, коли вони рівні та `False` – якщо ні.

Рекомендація до завдання: Не використовуйте оператор `==`

````python
import numpy as np

def equal_arrays(X, Y, Z):
    # Базовий випадок: коли один з масивів пустий
    if not X or not Y or not Z:
        return X == Y == Z
    
    # Рекурсивний випадок: коли всі масиви не пусті
    # Порівнюємо перші елементи з двох масивів
    if X[0] != Y[0] or X[0] != Z[0]:
        return False
    
    # Видаляємо перший елемент з кожного масиву
    X = X[1:]
    Y = Y[1:]
    Z = Z[1:]
    
    # Рекурсивно порівнюємо залишки масивів
    return equal_arrays(X, Y, Z)

# Використання функції
print(equal_arrays([1, 3, 3, 5], [1, 3, 3, 5], [1, 3, 3, 5]))  # True
print(equal_arrays([1, 3, 3, 5], [1, 3, 3, 5], [1, 3, 3, 7]))  # False
````

Завдання 8: Написати функцію яка приймає на вхід двовимірний масив, а повертає масив з діагональними елементами цієї матриці.

Наприклад: `[[1, 2], [3, 4]] -> [1 4]`

````python
import numpy as np

def diagonal_elements(arr):
    # Перетворюємо масив у numpy array
    arr = np.array(arr)
    
    # Отримуємо кількість рядків
    rows = len(arr)
    
    # Отримуємо кількість стовпців
    columns = len(arr[0])
    
    # Перевіряємо, чи є масив квадратним
    if rows != columns:
        raise ValueError("Масив має бути квадратним")
    
    # Отримуємо діагональні елементи
    diagonal_elements = arr.diagonal()
    
    return diagonal_elements

# Використання функції
print(diagonal_elements([[1, 2], [3, 4]]))  # [1 4]
````

Завдання 9: Написати функцію яка приймає на вхід двомірний масив, а повертає масив елементів які знаходяться на другорядній діагоналі. Приклад: `[[1, 2], [3, 4]] -> [2 3]`

````python
import numpy as np

def second_diagonal_elements(arr):
    arr = np.array(arr)
    n = arr.shape[0]
    second_diag_indices = np.arange(n-1, -1, -1)
    return arr[np.arange(n), second_diag_indices]

# Приклад використання:
array = [[1, 2], [3, 4]]
result = second_diagonal_elements(array)
print(result)  # Результат: [2 3]
````

$f_{4}(x) =  cos(x)$

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f1(x):
    return np.sin(x)/x

def f2(x):
    return np.cos(x)/(x**2 + 1)

def f3(x):
    return np.cos(x)**2 - np.sin(x)/2

def f4(x):
    return np.cos(x)

def f5(x):
    return np.sin(x)/(x**2 + 1)

# Generate x values
x = np.linspace(-10, 10, 1000)

# Plot the functions
plt.figure(figsize=(10, 6))
plt.plot(x, f1(x), label="$f_1(x) = sin(x)/x$")
plt.plot(x, f2(x), label="$f_2(x) = cos(x)/(x^2 + 1)$")
plt.plot(x, f3(x), label="$f_3(x) = cos^2(x) - sin(x)/2$")
plt.plot(x, f4(x), label="$f_4(x) = cos(x)$")
plt.plot(x, f5(x), label="$f_5(x) = sin(x)/(x^2 + 1)$")

# Add labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Show the plot
plt.show()
````
